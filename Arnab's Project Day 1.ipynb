* Write a function to download a file from internet

```python
import requests
import os
```

```python
basepath = "res/res"
if not os.path.exists(basepath):
    os.makedirs(basepath)
```

dir(requests)

```python
def download_a_file(url = 'http://google.com/favicon.ico', ):
    try:
        r = requests.get(url, allow_redirects=True)
        filename = url.split("/")[-1]
        open(os.path.join(basepath, filename), 'wb').write(r.content)
    except requests.exceptions.MissingSchema:
#         print(dir(e))
        print("Invalid URL: {}".format(url))
    except requests.exceptions.ConnectionError:
        print("Unable to connect to URL: {}".format(url))
```

```python
download_a_file("https://course.fast.ai/images/jupyter.png")
```

* Write a function to plot a Histogram from data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(3, 7, (10, 80))
data
```

```python
plt.plot(data) # plotting by columns
plt.show()
```

```python
def plot_histogram(data):
    fig, ax = plt.subplots(figsize=[10, 10])

    # set plot title
    ax.set_title("My Histogram")

    # set x-axis name
    ax.set_xlabel("X-Label")

    # set y-axis name
    ax.set_ylabel("Y-Label")

    N, bins, patches = ax.hist(data, bins = range(3,8), color="#777777") #initial color of all bins

    # Iterate through all histogram elements
    # each element in this interation is one patch on the histogram, where:
    # - bin_size - number of records in current bin
    # - bin - value of current bin (x-axis)
    # - patch - a rectangle, object of class matplotlib.patches.Patch
    # more details on patch properties: [visit this link][1]
    for bin_size, bin, patch in zip(N, bins, patches):
        facecolor = ["#00FF00", "#0000FF", "#FF0000", "#FFFF00"][bin-3]
        print(facecolor)
        patch.set_facecolor(facecolor)
        patch.set_label(str(bin))
    print(data)
    plt.show()
```

```python
data2 = np.random.randint(3, 7, (1, 10))
data2[0]
```

```python
plot_histogram(data2[0])
```
